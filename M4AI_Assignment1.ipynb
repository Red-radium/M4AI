This file contains tabular data. Each row of the CSV file corresponds to an *observation*/*subject*/*sample* from the dataset. Each column corresponds to a *feature*/*variable*/*characteristic* for each row.

To read tabular data stored as CSV files, we will rely on the **pandas** package. To load this package, we need to import it into the notebook:

```python
import pandas as pd
```

We can now read the CSV file using the *read_csv* method, and store it as a *pandas dataframe* object.

```python
tabular_data = pd.read_csv(filepath)
print(tabular_data)
```

We can then manipulate the dataframe using the built-in methods from the pandas package. Below are some examples.

```python
print(tabular_data.head())
age = tabular_data['Age']
print(age.mean())
# print(tabular_data.info())
# print(tabular_data.describe())
```

## Loading data: Image data (using OpenCV, Numpy and Matplotlib)

To read imaging data stored as *PNG*, *JPG* files (or other image formats), we will rely on the **OpenCV (CV2)** package. For manipulating and visualizing these images, we will rely on the **Numpy** and **Matplotlib** packages. To load this package, we need to import it into the notebook:

```python
# Import opencv module to read and display images
import cv2
from google.colab.patches import cv2_imshow
```

We need to once again specify the file path for the image before loading it using the imread method.

```python
image_file_path = '/content/drive/MyDrive/M4AI/parrot.png'
image = cv2.imread(image_file_path)
# image
```

To display the image, we use the **matplotlib** package:

```python
import matplotlib.pyplot as plt
```

We then use the *imshow* method to display the image.

```python
plt.imshow(image)
plt.axis('off')
```

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.axis('off')
```
