## Exercise 3: Plotting the Probability Mass Function (PMF) and Cumulative Distribution Function (CDF)

Consider the discrete random variable $X$ representing the number of heads in three coin flips.

1. Plot the PMF of $X$.
2. Plot the CDF of $X$.

```python
import matplotlib.pyplot as plt
import numpy as np

PMFX = [0,1,2,3]
PMFY = [1/8, 3/8, 3/8, 1/8]
plt.figure()
plt.scatter(PMFX, PMFY)
plt.title("PMF of X")
plt.xlabel("Number of heads")
plt.ylabel("Probability")
plt.grid(True)

CDFX = np.linspace(-2, 5, 100)
def CDF(x):
    if x<0:
        return 0
    elif 0<=x<1:
        return 1/8
    elif 1<=x<2:
        return 1/2
    elif 2<=x<3:
        return 7/8
    return 1

CDFY = [CDF(i) for i in CDFX]
plt.figure()
plt.plot(CDFX, CDFY)
plt.title("CDF of X")
plt.xlabel("Number of heads")
plt.ylabel("Cumulative probability")
plt.grid(True)

plt.show()
```

## Exercise 4: Plotting the Normal Distribution for Various Values of $\mu$ and $\sigma$

Define a *Python function* to generate plots of the PDF of the normal distribution for the following sets of parameters:

1. $\mu = 0$, $\sigma = 1$ (standard normal distribution)
2. $\mu = 0$, $\sigma = 2$
3. $\mu = 2$, $\sigma = 1$
4. $\mu = -2$, $\sigma = 0.5$

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, mu, sigma):
    # e = np.e
    first_term = 1/(sigma*np.sqrt(2*np.pi))
    second_term = np.e**(-((x-mu)**2)/(2*sigma*sigma))
    # print(second_term)
    return first_term*second_term

xs = np.linspace(-5,5,100)
y1 = [normal_distribution(x, 0, 1) for x in xs]
y2 = [normal_distribution(x, 0, 2) for x in xs]
y3 = [normal_distribution(x, 2, 1) for x in xs]
y4 = [normal_distribution(x, -2, 0.5) for x in xs]
ys = [y1, y2, y3, y4]

plt.figure()
plt.plot(xs, y1, label='μ=0, σ=1')
plt.plot(xs, y2, label='μ=0, σ=2')
plt.plot(xs, y3, label='μ=2, σ=1')
plt.plot(xs, y4, label='μ=-2, σ=0.5')
plt.title("Normal Distributions for various μ and σ")
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.grid(True)
plt.legend()


plt.show()
```
